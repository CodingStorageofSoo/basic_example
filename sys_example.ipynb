```python
import sys
```

```python
# 'sys.path' contains a list of directories that the interpreter will search in for the required module. 

sys.path
```

```python
# append() is a built-in function of sys module that can be used with path variable to add a specific path for interpreter to search.

sys.path.append ("address to add")
```

```python
# To remove the path 

sys.path.remove("address to remove")

# To remove the recent path 

sys.path.remove(sys.path[-1])
```
